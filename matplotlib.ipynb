```
import matplotlib.pyplot as plt
a= int(input("enter numbers for graph"))
b= int(input("enter numbers for graph"))
c= int(input("enter numbers for graph"))
e= int(input("enter numbers for graph"))
f= int(input("enter numbers for graph"))
g= int(input("enter numbers for graph"))
plt.bar([a,b,c],label="abc",color='orange',width=5,height=10)
plt.bar([e,f,g],label="efg",color='purple',width=5,height=10)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('graph')
plt.show()
```
